```python
import pandas as pd 
#Loads the train and test datasets.
test_data = pd.read_csv("Test.csv")
train_data = pd.read_csv("Train.csv")
```

```python
test_data
```

```python
train_data
```

```python
# Concatenate train and test datasets
combined_data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
combined_data
```

```python
# Check for missing values
print("\nMissing values in the combined dataset:")
print(combined_data.isnull().sum())
```

```python
# Impute missing values with the mean
combined_data.fillna(combined_data.mean(), inplace=True)
```

```python
# Display the dataset after handling missing values
print("\nDataset after handling missing values:")
print(combined_data.head())
```

```python
# Handle outliers using z-score
from scipy import stats
z_scores = stats.zscore(combined_data.select_dtypes(include=['int', 'float']))
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
combined_data = combined_data[filtered_entries]
```

# Exploratory Data Analysis (EDA)

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Summary statistics
print(combined_data.describe())
```

```python
# Distribution of target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='term_deposit_subscribed', data=combined_data)
plt.title("Distribution of Term Deposit Subscriptions")
plt.xlabel("Term Deposit Subscribed")
plt.ylabel("Count")
plt.show()
```

```python
# Pairplot to visualize relationships between numerical variables
numerical_cols = ['customer_age', 'balance', 'last_contact_duration', 'num_contacts_in_campaign', 
                  'days_since_prev_campaign_contact', 'num_contacts_prev_campaign']
sns.pairplot(combined_data[numerical_cols], diag_kind='kde')
plt.title("Pairplot of Numerical Variables")
plt.show()
```

```python
# Boxplot of numerical variables by target variable
plt.figure(figsize=(12, 8))
sns.boxplot(x='term_deposit_subscribed', y='customer_age', data=combined_data)
plt.title("Boxplot of Customer Age by Term Deposit Subscription")
plt.xlabel("Term Deposit Subscribed")
plt.ylabel("Customer Age")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Implement K-means clustering
kmeans = KMeans(n_clusters=3)
```

```python
# Select relevant features for clustering
features = ['customer_age', 'balance', 'last_contact_duration', 'num_contacts_in_campaign']
X = combined_data[features]
```

```python
#data scaling 
from sklearn.preprocessing import StandardScaler

# Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#dimensionality reduction 
from sklearn.decomposition import PCA

# Perform PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
# Clustering Algorithm Selection 
# K-means:Perform clustering using the selected algorithm
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X_pca)
```

```python
# Add cluster labels to the dataset
combined_data['cluster'] = kmeans.labels_
```

```python
# Visualize clustering results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=combined_data['cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering Results')
plt.show()
```
