```python
import pandas as pd 
#Loads the train and test datasets.
testdata = pd.read_csv("Test.csv")
traindata = pd.read_csv("Train.csv")
```

```python
#importing important libraries needed
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Combine the two datasets for preprocessing
combined_data = pd.concat([traindata, testdata], ignore_index=True)
combined_data
```

```python
# Handling Missing Values
imputer = SimpleImputer(strategy='mean')
combined_data[['customer_age', 'balance', 'last_contact_duration']] = imputer.fit_transform(combined_data[['customer_age', 'balance', 'last_contact_duration']])
```

```python
# Encode Categorical Variables
combined_data = pd.get_dummies(combined_data, columns=['job_type', 'marital', 'education', 'default', 'housing_loan', 'personal_loan', 'communication_type', 'month', 'prev_campaign_outcome'])
```

```python
# Split Data into Training and Testing Sets
X_train = combined_data.iloc[:traindata.shape[0], :-1]  # Features for training set
y_train = traindata['term_deposit_subscribed']          # Target variable for training set
X_test = combined_data.iloc[traindata.shape[0]:, :-1]   # Features for test set
```

```python
# Normalize Numerical Features
scaler = StandardScaler()
X_train[['customer_age', 'balance', 'last_contact_duration']] = scaler.fit_transform(X_train[['customer_age', 'balance', 'last_contact_duration']])
X_test[['customer_age', 'balance', 'last_contact_duration']] = scaler.transform(X_test[['customer_age', 'balance', 'last_contact_duration']])
```

```python
# Display preprocessed data
print("Preprocessed Training Data:")
print(X_train.head())
print("\nPreprocessed Testing Data:")
print(X_test.head())
```

```python
#EDA
#importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Summary statistics
print("\nSummary statistics of the combined dataset:")
print(combined_data.describe())
```

```python
# Check for missing values
print("\nMissing values in the combined dataset:")
print(combined_data.isnull().sum())
```

```python
# Visualize distribution of numerical features
numerical_features = ['customer_age', 'balance', 'last_contact_duration', 'days_since_prev_campaign_contact']
for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(combined_data[feature], kde=True, color='skyblue')
    plt.title(f'Distribution of {feature} in the combined dataset')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Visualize categorical features
categorical_features = ['job_type_admin', 'job_type_self-employed','marital_married', 'education_primary','education_tertiary' 'default_yes', 'housing_loan_no', 'personal_loan_no', 'communication_type_telephone', 'month_jun','month_jul','prev_campaign_outcome_success']

for feature in categorical_features:
    if feature in combined_data.columns:
        plt.figure(figsize=(10, 5))
        sns.countplot(data=combined_data, x=feature, palette='pastel')
        plt.title(f'Count of {feature}')
        plt.xlabel(feature)
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.show()
    else:
        print(f"'{feature}' not found in the dataset.")
```

```python
# Correlation heatmap (for numerical features only)
plt.figure(figsize=(10, 8))
sns.heatmap(combined_data[numerical_features].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap for the combined dataset')
plt.show()
```
