```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the data
#Loads the train and test datasets.
testdata = pd.read_csv("Test.csv")
traindata = pd.read_csv("Train.csv")
```

```python
testdata
```

```python
traindata
```

```python
data = pd.concat([traindata, testdata], ignore_index=True)
data
```

```python
data.dropna(inplace=True)

# Handling outliers/Remove outliers using z-score
from scipy import stats
z_scores = stats.zscore(data.select_dtypes(include=['int', 'float']))
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
data = data[filtered_entries]
```

```python
# Convert categorical variables to numerical using one-hot encoding
data = pd.get_dummies(data, columns=['job_type', 'marital', 'education', 'default', 'housing_loan', 'personal_loan', 'communication_type', 'month', 'prev_campaign_outcome'])

# Display the preprocessed dataset
print("\nPreprocessed Dataset:")
print(data.head())
```

```python
# Summary statistics
print("\nSummary Statistics:")
print(data.describe())
```

```python
# Check for missing values
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Distribution of target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='term_deposit_subscribed', data=data)
plt.title("Distribution of Term Deposit Subscriptions")
plt.xlabel("Term Deposit Subscribed")
plt.ylabel("Count")
plt.show()
```

```python
# Explore relationships between variables
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pairplot to visualize relationships between numerical variables
numerical_cols = ['customer_age', 'balance', 'last_contact_duration', 'num_contacts_in_campaign', 
                  'days_since_prev_campaign_contact', 'num_contacts_prev_campaign']
sns.pairplot(data[numerical_cols], diag_kind='kde')
plt.title("Pairplot of Numerical Variables")
plt.show()
```

```python
# Boxplot of numerical variables by target variable
plt.figure(figsize=(12, 8))
sns.boxplot(x='term_deposit_subscribed', y='customer_age', data=data)
plt.title("Boxplot of Customer Age by Term Deposit Subscription")
plt.xlabel("Term Deposit Subscribed")
plt.ylabel("Customer Age")
plt.show()
```

```python
# Visualize categorical features
categorical_features = ['job_type_blue-collar','marital_married', 'education_secondary','education_tertiary', 'default_no','default_yes', 'housing_loan_no', 'personal_loan_no', 'communication_type_telephone','communication_type_unknown', 'month_jan','month_jul','month_jun', 'prev_campaign_outcome_other','prev_campaign_outcome_success']
for feature in categorical_features:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=feature, hue='term_deposit_subscribed', data=data)
    plt.title(f'{feature.capitalize()} vs Term Deposit Subscription')
    plt.xlabel(feature.capitalize())
    plt.ylabel('Count')
    plt.legend(title='Term Deposit Subscribed', loc='upper right')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# Feature Selection
# Identify relevant features and attributes from the dataset for clustering.
features = ['customer_age', 'balance', 'last_contact_duration', 'num_contacts_in_campaign', 
            'days_since_prev_campaign_contact', 'num_contacts_prev_campaign']
```

```python
X = data[features]
```

```python
# Data Scaling
# Scale the data to ensure all features have equal influence in the clustering process.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dimensionality Reduction
pca = PCA(n_components=2)  
X_pca = pca.fit_transform(X_scaled)
```

```python
# Clustering Algorithm Selection 
# K-means:Perform clustering using the selected algorithm
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X_pca)
```

```python
# Add cluster labels to the dataset
data['cluster'] = kmeans.labels_
```

```python
# Visualize clustering results
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering Results')
plt.show()
```
